```python
# (Chris) importing library that does scraping from realtor.com
from homeharvest import scrape_property
from datetime import datetime
import pandas as pd
import numpy as np

# Generate filename based on current timestamp
#current_timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
#filename = f"HomeHarvest_{current_timestamp}.csv"

filename = 'LosAngelesHousingDataSoldLast30DaysChristopherKurdoghlian.csv'

# method with rows count for custom amount of rows
def get_dataset(filename, rows=None):
  max = 2000
  if rows != None:
    max = rows

  properties = scrape_property(
    location="Los Angeles, CA",
    listing_type="sold",  # or (for_sale, for_rent, pending)
    past_days=30,  # sold in last 30 days - listed in last 30 days if (for_sale, for_rent)
    limit = max

    # date_from="2023-05-01", # alternative to past_days
    # date_to="2023-05-28",
    # foreclosure=True
    # mls_only=True,  # only fetch MLS listings
  )

  #printing amount of properties
  print(f"Number of properties: {len(properties)}")

  # Export to csv
  properties.to_csv(filename, index=False)
  print(properties.head())



# calling method
# commented so we dont make a new file every time I do run all
get_dataset(filename, 5)
```

```python
# continuting to clean here (Chris)
# cleaning works like this
# doing some mode repacement with categorical type data
LA_data['nearby_schools'] = LA_data['nearby_schools'].replace(np.nan, LA_data['nearby_schools'].mode()[0])
LA_data['nearby_schools'].isna().sum()
```

```python
LA_data['neighborhoods'] = LA_data['neighborhoods'].replace(np.nan, LA_data['neighborhoods'].mode()[0])
LA_data['neighborhoods'].isna().sum()
```

```python
nan_count = LA_data.isna().sum().sum()
nan_count
#time to clean some more
```

```python
#lets see how many nulls we have for the latitude.
LA_data['latitude'].isna().sum()
```

```python
#lets see how many nulls we have for the longitude.
LA_data['longitude'].isna().sum()
```

```python
#latitude and longitude is gonna get replaced by the mean. I dont know if we will use it but
#it may come in handy for graphing. I would not want to train the data on the coordianes most likely.

LA_data['latitude'] = LA_data['latitude'].replace(np.nan, LA_data['latitude'].mean())
LA_data['longitude'] = LA_data['longitude'].replace(np.nan, LA_data['longitude'].mean())
```

```python
LA_data['longitude'].isna().sum()
```

```python
#now lets see lot_sqft
LA_data['lot_sqft'].isna().sum()
```

```python
#if the lot_sqft is null, replace it with the sqft value for that row.

LA_data['lot_sqft'] = LA_data['lot_sqft'].fillna(LA_data['sqft'])
```

```python
LA_data['lot_sqft'].isna().sum()
# this must be a listing that has both the log sqft and sqft as null. well drop it at the end.
```

```python
#if the estimated_value is null, replace it with the list_price value for that row.

LA_data['estimated_value'] = LA_data['estimated_value'].fillna(LA_data['list_price'])
LA_data['estimated_value'].isna().sum()
# we still have some nulls, we will drop em at the end.
#lastsold_date and sold_price have no nulls, as do list price and days on mls
```

```python
# noticed I skeipped price per square foot. Well come back to that.
```

```python
# replacing nans with median for the year built
LA_data['year_built'] = LA_data['year_built'].replace(np.nan, LA_data['year_built'].median())
```

```python
# now I will do sqft in a similar method we did lot sqft
LA_data['sqft'] = LA_data['sqft'].fillna(LA_data['lot_sqft'])

# now I will drop sqft that are na, since we did the replacing for the sqft and lot_sqft

LA_data = LA_data.dropna(subset=['sqft'])
LA_data = LA_data.dropna(subset=['lot_sqft'])
```

```python
# now I will do baths and beds
LA_data['Total Baths'] = LA_data['Total Baths'].replace(np.nan, LA_data['Total Baths'].median())
LA_data['beds'] = LA_data['beds'].replace(np.nan, LA_data['beds'].median())

# and the rest of the data should be full.
```

```python
nan_count = LA_data.isna().sum().sum()
nan_count
#lets see what columsn this is from.
```

```python
#I want to see how many nan values are in each column.

nan_counts = LA_data.isna().sum()
nan_counts
```

```python
#Ok, so a little more cleaning to do. as I mentioned, I am going to do the price_per_sqft calculation last.
#for the two list price and estimated values that must have both been null, I am going to replace it with the sold price.

LA_data['list_price'] = LA_data['list_price'].fillna(LA_data['sold_price'])
LA_data['estimated_value'] = LA_data['estimated_value'].fillna(LA_data['sold_price'])
```

```python
#for the days on mls, I am going to replace it with the median of the days on the mls.

LA_data['days_on_mls'] = LA_data['days_on_mls'].replace(np.nan, LA_data['days_on_mls'].median())
```

```python
# I am going to drop that one row that has a null value for the street.

LA_data = LA_data.dropna(subset=['street'])
```

```python
# finally, I will replace nan values on price_per_sqft by a calculation of the price per square foot, which would be the sold price divided by the sqft.

LA_data['price_per_sqft'] = LA_data['price_per_sqft'].fillna(LA_data['sold_price'] / LA_data['sqft'])
```

```python
#now we check again.
nan_counts = LA_data.isna().sum()
nan_counts
```

```python
LA_data = LA_data.dropna(subset=['agent_name'])
```

```python
# prompt: I want to visualize this attributes here with a visual that is most appropriate. sold_price, sqft, neighborhood, city, style

import matplotlib.pyplot as plt
import seaborn as sns

lower_bound = LA_data['sqft'].quantile(0.05)
upper_bound = LA_data['sqft'].quantile(0.95)
LA_data_filtered = LA_data[(LA_data['sqft'] >= lower_bound) & (LA_data['sqft'] <= upper_bound)]

# Create a box plot to visualize the relationship between neighborhood and sold price.
plt.figure(figsize=(15, 6))
sns.boxplot(x='city', y='sold_price', data=LA_data_filtered)
plt.xticks(rotation=90)
plt.title('Sold Price Distribution by City')
plt.xlabel('City')
plt.ylabel('Sold Price')
plt.show()

# Create a scatter plot to visualize the relationship between sqft and sold price.
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sqft', y='sold_price', data=LA_data_filtered)
plt.title('Sold Price vs. Square Footage')
plt.xlabel('Square Footage')
plt.ylabel('Sold Price')
plt.show()

# Create a bar plot to visualize the average sold price for each style.
average_price_by_style = LA_data_filtered.groupby('style')['sold_price'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=average_price_by_style.index, y=average_price_by_style.values)
plt.xticks(rotation=90)
plt.title('Average Sold Price by Property Style')
plt.xlabel('Property Style')
plt.ylabel('Average Sold Price')
plt.show()


# Create a count plot to visualize the number of properties in each city.
plt.figure(figsize=(10, 6))
sns.countplot(x='city', data=LA_data_filtered)
plt.xticks(rotation=90)
plt.title('Number of Properties by City')
plt.xlabel('City')
plt.ylabel('Number of Properties')
plt.show()
```

```python
# prompt: sqaure footage vs sold_price scatter plot but remove are properties of type LAND. Remove extrme outliers via quantiles. change scale to 1e6 for y axis. upper bound y axis to go to 4

# Filter out properties of type LAND
LA_data_filtered = LA_data[LA_data['style'] != 'Land']

# Remove extreme outliers in sqft using quantiles
lower_bound_sqft = LA_data_filtered['sqft'].quantile(0.05)
upper_bound_sqft = LA_data_filtered['sqft'].quantile(0.95)
LA_data_filtered = LA_data_filtered[(LA_data_filtered['sqft'] >= lower_bound_sqft) & (LA_data_filtered['sqft'] <= upper_bound_sqft)]

# Remove extreme outliers in sold_price using quantiles
lower_bound_price = LA_data_filtered['sold_price'].quantile(0.05)
upper_bound_price = LA_data_filtered['sold_price'].quantile(0.95)
LA_data_filtered = LA_data_filtered[(LA_data_filtered['sold_price'] >= lower_bound_price) & (LA_data_filtered['sold_price'] <= upper_bound_price)]

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(LA_data_filtered['sqft'], LA_data_filtered['sold_price'])
plt.xlabel('Square Footage (sqft)')
plt.ylabel('Price ($ Millions)')
plt.title('Price vs. Square Footage (Excluding Land, Mobile, and extreme outlier Properties)')
plt.yscale('linear')
plt.ylim(0, 4000000)
plt.yticks([0, 1000000, 2000000, 3000000, 4000000], ['0', '1', '2', '3', '4'])
plt.show()
```
